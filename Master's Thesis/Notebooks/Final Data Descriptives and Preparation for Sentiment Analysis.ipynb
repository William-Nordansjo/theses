```python
import pandas as pd

# Setup
file_path = "/Volumes/T7/Thesis data end product/nasdaq_exteral_data.csv"
chunk_size = 500_000
stocks = ["AAPL", "MSFT", "GOOGL", "AMZN", "NVDA", "META", "TSLA"]

# Step 1: Convert index lists to sets for fast lookup
stock_index_sets = {
    stock: set(globals()[f"df_index_{stock}"]["index"]) for stock in stocks
}

# Step 2: Prepare empty collectors
stock_chunks = {stock: [] for stock in stocks}

# Step 3: Process in chunks
reader = pd.read_csv(file_path, chunksize=chunk_size, usecols=["Date", "Article_title", "Article"])

for i, chunk in enumerate(reader):
    print(f"ðŸ”„ Processing chunk {i + 1}...")

    chunk = chunk.reset_index()  # Make the row index into a column called 'index'

    for stock in stocks:
        index_set = stock_index_sets[stock]
        matched = chunk[chunk["index"].isin(index_set)]
        if not matched.empty:
            stock_chunks[stock].append(matched[["index", "Date", "Article_title", "Article"]])

# Step 4: Concatenate into final stock-specific DataFrames
for stock in stocks:
    df_final = pd.concat(stock_chunks[stock], ignore_index=True)
    globals()[f"df_{stock.lower()}_articles"] = df_final
    print(f"âœ… Created df_{stock.lower()}_articles with {len(df_final)} rows")
```

```python
import pandas as pd
import re

import re
import pandas as pd

def clean_article(text):
    if pd.isna(text):
        return ""

    # Fix escaped newlines and other formatting
    text = re.sub(r'\\[ntr]', ' ', text)

    # Fix squished words (add space between lowercase-uppercase boundaries if not already present)
    text = re.sub(r'(?<=[a-z])(?=[A-Z])', ' ', text)

    # Remove any remaining HTML tags (if present)
    text = re.sub(r'<.*?>', '', text)

    # Remove non-ASCII characters (e.g., weird unicode)
    text = text.encode('ascii', errors='ignore').decode()

    # Normalize spacing
    text = re.sub(r'\s+', ' ', text).strip()

    return text


# List of your article DataFrames
article_dfs = [
    df_aapl_articles,
    df_msft_articles,
    df_googl_articles,
    df_amzn_articles,
    df_nvda_articles,
    df_meta_articles,
    df_tsla_articles
]

# Apply cleaning to each DataFrame
for df in article_dfs:
    df["Article"] = df["Article"].apply(clean_article)
```
